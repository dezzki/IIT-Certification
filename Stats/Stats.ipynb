```python
#outliers : The extreem low or high datapoints/Values in data.

# Outlier Detection: We need to detect the outliers with box plot.
# Outlier Removal : We set an IQR( Inter quartile range) and remove the outliers.
```

```python
import matplotlib.pyplot as plt
```

```python
age = [25,23,38,39,65,73,55,4]
plt.boxplot(age)
plt.show()
```

```python
age = [25,23,38,39,65,73,55,180,-55]
plt.boxplot(age)
plt.show()
```

```python
# Outlier Removal - IQR Method
# IQR - Inter Quartile Range?
```

```python
# Step 1 : Calculate Q1 & Q3
# Step 2 : Calculate IQR -> Q3 - Q1
# Step 3 : Lower Bound -> LB = Q1 - 1.5*IQR 
# Step 4 : Upper Bound -> UB = Q3 + 1.5*IQR
# Step 5 : Take Values only greater than LB and smaller than UB.
```

```python
age = [25,23,38,39,65,73,55,180,-55]

Q1 = np.percentile(age,25)
Q3 = np.percentile(age,75)
print(Q1,"  ",Q3)

IQR = Q3 - Q1
print(IQR)
```

```python
lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR
print(lower_bound, upper_bound)

normal_age = []
for i in age:
    if (i < lower_bound) or (i > upper_bound):
        print(i, "is an outlier")
    else:
        normal_age.append(i)

print(normal_age)
```

```python
plt.boxplot(normal_age)
plt.show()
```

```python
# KURTOSIS
    # Where the data peaked with heavy tails and packed with light tails.
    # How heavy/light tails of distribution are

# K = 3 , Normal Distribution, Kurtosis is normal, Mesorkurtik
# K > 3 , It has a sharp peak, Kurtosis has heavy tails, Laptokurtik 
# K < 3 , It has flat peak, Kurtois is Light tailed,  

# Section A - Most students scored avg marks.
# Section B - Some students scored very high / some scored very low.

# Section B has more outliers -> Higher kurtosis.
```

```python
# Heavy tail data

heavy_tail = np.concatenate([
    np.random.normal(50,5,1000),
    [10,15,100,10]
])

print(np.random.normal(50,5,1000))

plt.figure(figsize=(8,5))
sns.histplot(heavy_tail, kde="true")

plt.title("Heavy Tail")
plt.xlabel("Values")
plt.ylabel("Frequency")

plt.show()
```

```python
# Types of Kurtosis :
    
    # Mesokurtik ( Normal , Kurtosis = 3), noraml distribution, moderate/normal tail,
    # Laptokurtik ( High , Kurtosis > 3) , Shark peak + Heavy tails
    # Platikurtik ( Low , Kurtosis < 3 ) , Flat distribution, evenly spread
```

```python
# Probability

    # It is a measure of how likely an event is going to occur/possibilty of an event.
    # It always lies between 0 and 1.

    # 0 -> no chances (will never happen).
    # 1 -> The event will definately happen.
    # 0-1 -> May happen.
```

```python
# Experiment : any process or any task
# ex - Tossing a coin, rolling a dice.

# Event = outcome in which we are interested
# tossing a coin = E1 -> H, E2 -> T.
```

```python
# Types of events

    # Simple : It contains only one soecific outcome / One exact value

        # ex. -> Tossing a coin and getting tail.
        # rolling a dice and getting 3

    # Compound event = Contains more than one possible outcome.
        # Ex -> Rolling a dice and getting eve nnumbers( 2,4,6 ) / Odd numbers( 1,3,5 )

# Probability formula = No of fav. outcomes/Total number of outcomes.
```

```python
# Types of Probability

    # Marginal Probability
    # Joint Probability
    # Conditional Probability


# Marginal Probability

    # It is the posibilty of one independent event ignoring everything
    # ex = Tossing a coin.

    # Event A -> Getting Heads
    # possible outcomes -> H, T
    # Fav outcomes -> H

    # p(A) -> 1/2
    # We are not considering any other event other than getting Head.


# Joint Probability

    # Probability of 2 events happening together.
    # Event(A) -> First coin getting a Head.
    # Event(B) -> Second coin getting a Head.

    # outcomes -> {HH,TT,HT,TH}
    #P(A and B)-> 1/4 or .25


# Conditional Probability

    #Probability of an event happening given that another event haas already occured.

    #Total cards = 52
    # Event A -> card is a King
    # Event B -> card is a face card 

    # What is the posibility that the card drawn is a king given that it is a face card.
    # p -> 4/12 -> 1/3.
```
