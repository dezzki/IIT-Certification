```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('BBData.csv')
df = pd.DataFrame(data)
df.head(5)
```

```python
# print no of rows and columns

print(df.shape)
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df.sample(7)
```

```python
df = df.drop(columns = 'index')
df.isnull().sum()
```

```python
df = df.dropna(subset=['product'])
```

```python
df['brand'] = df['brand'].fillna("Unknown")
df['brand'].info()
```

```python
df['description'] = df['description'].fillna("Not Available")
```

```python
df.isnull().sum()
```

```python
df1 = df['rating'] = df.groupby(['brand' , 'category'])['rating'].transform(lambda x:x.fillna(x.mean))
```

```python
df1
```

```python
print(df.duplicated().sum())
```

```python
df.drop_duplicates(inplace= True)
```

```python
df.to_csv("EDA File Cleaned.csv")
```
